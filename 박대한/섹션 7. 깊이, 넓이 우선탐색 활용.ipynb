## 7. 송아지찾기

```python
import sys
#sys.stdin=open("input.txt", "r")
from collections import deque
s,e = map(int,input().split())
max=10000
ch=[0]*(max+1)
dis=[0]*(max+1)
ch[s]=1
dis[s]=0
dq=deque()
dq.append(s)
while dq : 
    now=dq.popleft()
    if now == e : 
        break
    for next in [now-1,now+1,now+5] : 
        if next>=1 and next<=max and ch[next]==0: 
            dq.append(next)
            ch[next]=1
            dis[next]=dis[now]+1
print(dis[e])
```

## 14. 안전영역

```python
import sys
#sys.stdin=open("input.txt", "r")
from collections import deque
if __name__ == '__main__' : 
    n = int(input())
    a1=[]
    minimum,maximum = float('inf'), -float('inf')
    for i in range(n) : 
        b = list(map(int,input().split()))
        a1.append(b)
        for j in b :
            if j < minimum : 
                minimum = j
            if j > maximum : 
                maximum = j
    dq = deque()
    dx = [-1,0,1,0]
    dy = [0,-1,0,1]
    res=[]
    for limit in range(minimum,maximum) : 
        ch=[[0]*n for _ in range(n)]
        cnt=0
        for i in range(n) : 
            for j in range(n) : 
                if a1[i][j] > limit and ch[i][j] == 0 : 
                    ch[i][j] = 1
                    dq.append((i,j))
                    while dq :
                        now = dq.popleft()
                        x,y = now[0],now[1]
                        for k in range(4) : 
                            newx, newy = x+dx[k], y+dy[k]
                            if 0<=newx<=n-1 and 0<=newy<=n-1 and a1[newx][newy] > limit and ch[newx][newy] == 0 : 
                                ch[newx][newy] = 1
                                dq.append((newx,newy))
                    cnt += 1
        res.append(cnt)
    print(max(res))
```

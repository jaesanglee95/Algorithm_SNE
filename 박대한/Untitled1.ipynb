```python
# 다른 사람 풀이 참고
from itertools import combinations
def solution(info,query) : 
    answer = []
    db={}
    for i in info : 
        temp = i.split()
        conditions=temp[:-1]
        score=int(temp[-1])
        for n in range(5) : # n : 몇개의 조건을 -처리 할거냐
            combi = list(combinations(range(4), n))
            for c in combi:
                t_c = conditions.copy()
                for v in c:
                    t_c[v] = '-'
                changed_t_c='/'.join(t_c)
                if changed_t_c in db : 
                    db[changed_t_c].append(score)
                else : 
                    db[changed_t_c]=[score]
                    
    for value in db.values() : 
        value.sort()
        
    for q in query :       
        qry = [i for i in q.split() if i != 'and']
        qry_cnd = '/'.join(qry[:-1])
        qry_score = int(qry[-1])
        if qry_cnd in db: 
            data = db[qry_cnd]
            if len(data) > 0 : 
                start, end = 0, len(data)
                while start != end and start != len(data):
                    if data[(start + end) // 2] >= qry_score:
                        end = (start + end) // 2
                    else:
                        start = (start + end) // 2 + 1
                answer.append(len(data) - start)
        else:
            answer.append(0)
    return answer
```

```python
# 키패드 누르기 100점.
def solution(numbers, hand):
    answer = ''
    nowL=(3,0)
    nowR=(3,2)
    for n in numbers : 
        if n == 1 or n==4 or n==7 : 
            answer += 'L'
            nowL = (n//3,0)
        elif n==3 or n==6 or n==9 : 
            answer += 'R'
            nowR = ((n//3)-1,0)
        else : 
            if n == 0 : 
                now = (3,1)
            else : 
                now = (n//3,1)
            gapL = abs(now[0]-nowL[0]) + abs(now[1] - nowL[1])
            gapR = abs(now[0]-nowR[0]) + abs(now[1] - nowR[1])
            if gapL > gapR : 
                answer += 'R'
                nowR = now
            elif gapL < gapR : 
                answer += 'L'
                nowL = now
            else : 
                if hand == 'right' : 
                    answer += 'R'
                    nowR = now
                else : 
                    answer += 'L'
                    nowL = now
    return answer
```

```python
# 보석 쇼핑
# 카카오문제해설, 다른 사람 풀이보면서 풀었음.
def solution(gems) : 
    uniquenum=len(list(set(gems)))
    answer=[0,len(gems)-1]
    maps={}
    maps[gems[0]]=1
    lt,rt=0,0
    while 0<=lt<len(gems) and 0<=rt<len(gems) : 
        if len(maps) < uniquenum : 
            rt+=1
            if rt==len(gems) :
                break
            if gems[rt] in maps :
                maps[gems[rt]]+=1
            else : 
                maps[gems[rt]]=1
        else : 
            if rt-lt < answer[1]-answer[0] : 
                answer = [lt,rt]
            maps[gems[lt]]-=1
            if maps[gems[lt]]==0 : 
                del maps[gems[lt]]
            lt+=1
    answer[0]+=1
    answer[1]+=1
    return answer
```
